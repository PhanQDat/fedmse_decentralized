```python
import pandas as pd
import numpy as np
import os
from datetime import datetime, timedelta
import random

data_path_folder = '../Active Wiretap/'
data_path_dataset = os.path.join(data_path_folder, 'Active_Wiretap_dataset.csv')
data_path_label = os.path.join(data_path_folder, 'Active_Wiretap_labels.csv')
# Đọc dữ liệu
df_dataset = pd.read_csv(data_path_dataset)
df_labels = pd.read_csv(data_path_label)
```

```python
df_dataset
```

```python
# Gắn nhãn từ cột 'x' trong labels vào cột 'label' của dữ liệu gốc
df_dataset['label'] = df_labels['x']
df_dataset
```

```python
data_normal = df_dataset[df_dataset['label'] == 0]
data_abnormal = df_dataset[df_dataset['label'] == 1]
```
